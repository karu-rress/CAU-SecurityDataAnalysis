## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
```

## PLOT

```python
random_n = randn(50).cumsum()
plt.plot(random_n)
```

```python
plt.plot(random_n, color='r', linestyle='--', linewidth=3)
```

```python
plt.plot(random_n, color='b', linestyle='dashed', linewidth=1, marker='o', label='Default')
```

```python
# steps-post, stpes-pre, steps-mid, steps
plt.plot(random_n, color='b', linestyle='dashed', linewidth=1, marker='o', label='steps-post', drawstyle='steps-post')
```

```python
plt.plot(random_n, color='b', linewidth=1, label='Default')
plt.plot(random_n, color='red', linestyle='dashed', linewidth=1, label='steps-post', drawstyle='steps-post')
plt.legend(loc='best')
```

```python
plt.plot(random_n, 'k--')
```

```python
plt.plot(random_n, 'go', linewidth=2) 
```

```python
plt.plot(random_n, 'b+', linewidth=5) 
```

# Subplots

```python
fig = plt.figure()
# row, column, plot no.
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

_ = ax1.plot(random_n, 'k--')
ax2.plot(random_n, 'go', linewidth=2) 
_ = ax3.scatter(x, random_n, color='r', linewidth=0.1)
ax4.hist(random_n, bins=20, color='b', alpha=0.5)
```

```python
# using axes with index
fig2, axes = plt.subplots(2,2)
_ = axes[0,0].plot(random_n, 'k--') 
axes[0,1].plot(random_n, 'go', linewidth=2) 
_ = axes[1,0].scatter(x, random_n, color='r', linewidth=0.1)
axes[1,1].hist(random_n, bins=20, color='b', alpha=0.5)
```

```python
fig3, axes = plt.subplots(2, 2, sharex=True, sharey=True)

color_list = ['r', 'b', 'k', 'y']

k = 0
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500), bins=50, color=color_list[k], 			alpha=0.5, label='legend '+color_list[k])
        axes[i,j].legend()
        k += 1

#  
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label ='one')
ax.plot(randn(1000).cumsum(), 'b--', label='two')
ax.plot(randn(1000).cumsum(), 'y.', label='three') # label  , label='_nolegend_'   

ax.set_xticks([0, 250, 500, 750, 1000]) #  x 
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title("Visualization practice for plot")
ax.set_xlabel("Stages")
ax.set_ylabel("y axis")
ax.set_xlim(-50, 1000) #    
# ax.set_ylim(-50, 40)   ...
ax.legend(loc='best')
```

## Examples

```python
from datetime import datetime
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('data/3_spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k--')

crisis_date = [(datetime(2007, 10, 11), "Peak of bull market"),
             (datetime(2008, 3, 12), "Bear Stearns Fails"),
             (datetime(2008, 9, 15), 'Lehman Bankruptcy')]

for date, label in crisis_date:
    ax.annotate(label, xy=(date, spx.asof(date)+50), xytext=(date, spx.asof(date)+200), 
                arrowprops=dict(facecolor='black'), horizontalalignment='left', verticalalignment='top')

#  string  convert .
# numpy datetime   .
import numpy as np
ax.set_xlim(np.array(['2007-01-01', '2011-01-01'], dtype='datetime64'))
ax.set_ylim(600, 1800)

ax.set_title('Important dates in 2008-2009 financial crisis')
```
