## 카테고리화

```python
df_count = df.groupby(['type']).count()
display(df_count)
print('len(df_count):', len(df_count))
print('len(df):', len(df))
```

```python
print(len(df[df["type"]=="CASH_OUT"]))
display(df.groupby(["type"]).mean(numeric_only=True))
display(df.groupby(["type"]).sum())
display(df.groupby(["type", "isFraud"]).mean(numeric_only=True))
display(df.groupby(["type", "isFraud"]).count())
```

## Correlation and Covariance

```python
print(df.columns)
print()
# corr : [-1. 1]
print(df['oldbalanceOrg'].corr(df['newbalanceOrig']))

# cov
print(df.oldbalanceOrg.cov(df.newbalanceOrig))
```

```python
# 상관분석
display(df.corr(numeric_only=True))

# 공분산 분석
display(df.cov(numeric_only=True))
```

```python
df.corrwith(df.amount, numeric_only=True)
```

```python
# N/A 10개 날아감
df_copy.dropna(inplace=True)
```

```python
df.dropna(how='all')
```

```python
df.iloc[:2, :] = None
df.dropna(how="all")
```

```python
df["all_NA"] = None
df.head()
```

```python
# 좌에서 우로
df.dropna(axis=1, how='all')
```

```python
print(len(df.columns))
display(df.dropna(thresh=2))
display(df.dropna(thresh=11))
display(df)
```

```python
display(df.dropna(axis=1, thresh=11))
display(df.fillna(0))
```

```python
# 결측값을 평균으로 채우기
print(df['amount'].mean())
display(df.fillna({"isFraud":0, "amount":df['amount'].mean()}))


df.iloc[196:200,:] = None
display(df)

# front fill
df.fillna(method="ffill")
display(df)

df.fillna(method="ffill", limit=2)
```
