### 각각의 칼럼들끼리는 뚜렷한 상관관계는 없는 것 같음. amount와 newbalanceDest 사이 약간의 상관관계가 보이긴 하지만 송금, 지불 등 오간 돈의 양이 커질 수록 대상 잔고가 높아지는 효과 같음.

```python
df1 = df.copy()
```

```python
df1['realAmountOrig'] = abs(df1['newbalanceOrig'] - df1['oldbalanceOrg'])
df1
```

```python
df1['realAmountDest'] = abs(df1['newbalanceDest'] - df1['oldbalanceDest'])
df1
```

```python
df.type.unique()
```

```python
df_minus = df1[df1['type'].isin(['PAYMENT', 'TRANSFER', 'CASH_OUT', 'DEBIT'])] # 정상적으로 수행될 때 Orig 잔고가 줄어드는 경우
df_plus = df1[df1['type'] == 'CASH_IN'] # 정상적으로 수행될 때 Orig 잔고가 늘어나는 경우
```

```python
df_minus
```

```python
len(df_plus[df_plus['amount'].round(2) != df_plus['realAmountOrig'].round(2)])
```

```python
df_plus[df_plus['amount'].round(2) != df_plus['realAmountOrig'].round(2)][['amount', 'realAmountOrig']]
```

```python
temp1 = abs(df_plus['amount'].round(2) - df_plus['realAmountOrig'].round(2)).round(2)
temp1
```

```python
temp1 = pd.DataFrame(temp1, columns = ['difference'])
temp2 = temp1[temp1['difference'] != 0]
```

```python
(temp1['difference'].round(2)).unique()
```

```python
ax = sns.countplot(x = 'difference', data = temp1)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=10)
```

```python
ax = sns.countplot(x = 'difference', data = temp2)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=10)
```

```python
temp3 = temp2[temp2['difference'] != 0.01]
```

```python
ax = sns.countplot(x = 'difference', data = temp3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=10)
```

```python
count = []
count.append(len(temp1[temp1['difference'] == 0]))
count.append(len(temp1[temp1['difference'] == 0.01]))
count.append(len(temp1[temp1['difference'] > 0.02]))
count
```

```python
df_count = pd.DataFrame(count, index = ['0', '0.01', '0.02+'], columns = ['count'])
df_count
```

```python
plt.pie(df_count['count'], labels = ['0', '0.1', '0.2+'])
```
