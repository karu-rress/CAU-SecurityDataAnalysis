## 2.3. Check outliers using IQR (on train set)

```python
df.describe()
```

```python
df_numeric = df.select_dtypes(include=['int64', 'float64'])
Q1 = df_numeric.quantile(0.25)
Q2 = df_numeric.quantile(0.5)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
IQR_DELTA = abs((df_numeric - Q2) / IQR)
IQR_DELTA
```

```python
# Fill NaN and Inf values with 0
IQR_DELTA = IQR_DELTA.replace([np.inf, -np.inf, np.nan], 0)
IQR_DELTA
```

```python
IQR_MAX = IQR_DELTA.max(axis=1)
IQR_MAX
```

```python
for i in range(1, 30):
    df_outlier = df_numeric[(IQR_MAX > i) & ~df_numeric['Success']]
    print(f"Threshold: {i}-IQR, # of outliers: {len(df_outlier)}, % of outliers: {round(len(df_outlier) / len(df) * 100, 1)}%")
```

Given the success rate is ~28%, threshold of 15-IQR seems reasonable.

```python
IQR_OUTLIER_THRESHOLD = 15
df_outlier = df_numeric[IQR_MAX > IQR_OUTLIER_THRESHOLD]
df_outlier
```

```python
df = df.drop(df_outlier.index)
```

# 7. Preparing train & test set

Using previously marked `is_train` column, we will split the dataset into train and test set.

```python
target_column = 'FundedRate'

X_train = X[X['is_train']].drop(['is_train'], axis=1)
y_train = y[X['is_train']]
y_train_logi = y_logi[X['is_train']]

X_test = X[~X['is_train']].drop(['is_train'], axis=1)
y_test = y[~X['is_train']]
y_test_logi = y_logi[~X['is_train']]
```

# 8. Feature Scaling

Scale all numeric variables to have a mean of 0 and a standard deviation of 1.
(One-hot encoded variables, which is type of uint8, is excluded)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaling_columns = X.select_dtypes(include=['int64', 'float64']).columns

scaler.fit(X_train[scaling_columns])
X_train[scaling_columns] = scaler.transform(X_train[scaling_columns])
X_test[scaling_columns] = scaler.transform(X_test[scaling_columns])
```

## 8.a. Upon trials and errors, it is concluded that the range of some variables are too wide, we may need non-linear transformation to make the distribution more even.

## `Backers`, `Goal`, `Funded`

```python
from sklearn.preprocessing import PowerTransformer

transformer = PowerTransformer()
transforming_columns = ['Backers', 'Goal']

transformer.fit(X_train[transforming_columns])
X_train[transforming_columns] = transformer.transform(X_train[transforming_columns])
X_test[transforming_columns] = transformer.transform(X_test[transforming_columns])
```

# 10. Model: Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
pd.DataFrame({'column': X_train.columns, 'coef': model.coef_.squeeze()})
```

```python
print(f"Train set score: {model.score(X_train, y_train)}")
print(f"Test set score: {model.score(X_test, y_test)}")
```
