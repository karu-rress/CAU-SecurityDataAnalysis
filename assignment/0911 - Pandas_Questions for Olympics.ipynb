### 0. Read and show the file content.
- Please show the cleansing work here if you use the first file.

```python
import pandas as pd

# Set first row as header & refine header
df = pd.read_csv('../data/0911_2_olympics.csv', header=1)
df.rename(columns={'Unnamed: 0':'Country'}, inplace=True)
df.columns = df.columns.str.replace('№', 'No.', )\
    .str.replace('01 !', 'Gold').str.replace('02 !', 'Silver').str.replace('03 !', 'Bronze')

# remove last row
df = df[:-1]

# Extract country name in first col
df['Country'] = df['Country'].str.extract(r'([\w ]+)')
df
```
